```python
import pandas as pd


label_data = cmumosei['All Labels'].data


data = []

for segment_id, segment_group in label_data.items():
    if segment_id not in target_segment_ids:
        continue

    score_array = segment_group['features'][:]
    score = score_array[0][0]

    if score > 0.3:
        label = 2
    elif score < -0.3:
        label = 0
    else:
        label = 1

    data.append({
        "segment_id": segment_id,
        "sentiment_score": score,
        "sentiment_label": label
    })

df_labels = pd.DataFrame(data)
print(df_labels.head())
```

```python
import numpy as np
import pandas as pd

glove_data = cmumosei['glove_vectors'].data

glove_vectors = []

for segment_id, group in glove_data.items():
    if segment_id not in target_segment_ids:
        continue
    try:
        vectors = group['features'][:]
        if vectors.shape[0] == 0:
            continue
        avg_vector = np.mean(vectors, axis=0)
        glove_entry = {"segment_id": segment_id}
        for i, val in enumerate(avg_vector):
            glove_entry[f"glove_{i}"] = val

        glove_vectors.append(glove_entry)
    except Exception as e:
        print(f"Skipping {segment_id}: {e}")
        continue

df_glove_avg = pd.DataFrame(glove_vectors)
print(df_glove_avg.head())
print(len(df_glove_avg))
```

```python
import numpy as np

segment_texts = {}

for segment_id, group in cmumosei['CMU_MOSEI_TimestampedWords'].data.items():
    if segment_id not in target_segment_ids:
        continue
    try:
        features = group['features'][:]
        print(f"\nSegment: {segment_id}")

        decoded_words = []
        for w in features:
            if isinstance(w, (list, np.ndarray)) and len(w) > 0 and isinstance(w[0], bytes):
                try:
                    decoded = w[0].decode('utf-8').strip()
                    print(f"Decoded word: {repr(decoded)}")
                    if decoded and decoded.lower() != 'sp':
                        decoded_words.append(decoded)
                except Exception as decode_err:
                    print(f"{w[0]}: {decode_err}")

        print(f"Decoded words list: {decoded_words}")

        sentence = ' '.join(decoded_words).strip()
        print(f"Decoded sentence: \"{sentence}\"")

        if sentence:
            segment_texts[segment_id] = sentence

    except Exception as e:
        print(f"{segment_id} {e}")
```

```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

bert_vectors = []

for segment_id, sentence in segment_texts.items():
    try:
        inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)
        new_inputs = {}
        for k, v in inputs.items():
            new_inputs[k] = v.to(device)
        inputs = new_inputs

        with torch.no_grad():
            outputs = model(**inputs)
            pooler_output = outputs.pooler_output.squeeze().cpu().numpy()

        bert_entry = {"segment_id": segment_id}
        for i, val in enumerate(pooler_output):
            bert_entry[f"bert_{i}"] = val

        bert_vectors.append(bert_entry)
    except Exception as e:
        print(f"Skipping {segment_id}: {e}")
```

```python
train_df = df_full[df_full['split'] == 'train']
val_df   = df_full[df_full['split'] == 'val']
test_df  = df_full[df_full['split'] == 'test']
```

```python
drop_cols = ['segment_id', 'sentiment_score', 'sentiment_label', 'split']
X_train = train_df.drop(columns=drop_cols)
y_train = train_df['sentiment_label']

X_test = test_df.drop(columns=drop_cols)
y_test = test_df['sentiment_label']
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32).to(device)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32).to(device)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long).to(device)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long).to(device)


class MLPClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim=128, output_dim=3):
        super(MLPClassifier, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.net(x)

model = MLPClassifier(input_dim=X_train.shape[1]).to(device)


criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
epochs = 20


for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")


model.eval()
with torch.no_grad():
    logits = model(X_test_tensor)
    y_pred = torch.argmax(logits, dim=1).cpu().numpy()
    y_true = y_test_tensor.cpu().numpy()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=["Negative", "Neutral", "Positive"]))
```
